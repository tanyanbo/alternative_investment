```python
from scipy.stats import rankdata
import scipy as sp
import numpy as np
import pandas as pd
import datetime
```

```python
df = pd.read_csv('./alldata.csv', encoding="ISO-8859-1")
```

```python
df['Trddt'] = pd.to_datetime(df['Trddt'], format='%Y-%m-%d')
df
```

```python
df1 = pd.read_csv('FI_T1.csv', names=['code', 'date', 'type', 'current', 'quick', 'cash', 'asset_liabilities'], skiprows=1)
df1['date'] = pd.to_datetime(df1['date'], format='%Y-%m-%d')
df1 = df1[df1.date >= datetime.datetime(2010,1,1)].fillna(0)
df1['total'] = df1.apply(lambda row: row.current + row.quick + row.cash + row.asset_liabilities, axis=1)
df1
```

```python
df2 = pd.read_csv('FI_T4.csv', names=['code', 'date', 'type', 'ar', 'inventory', 'total_assets'], skiprows=1)
df2['date'] = pd.to_datetime(df2['date'], format='%Y-%m-%d')
df2 = df2[df2.date >= datetime.datetime(2010,1,1)].fillna(0)
df2['total'] = df2.apply(lambda row: row.ar + row.inventory + row.total_assets, axis=1)
df2
```

```python
df3 = pd.read_csv('FI_T5.csv', names=['code', 'date', 'type', 'roa', 'roe', 'gross', 'net'], skiprows=1)
df3['date'] = pd.to_datetime(df3['date'], format='%Y-%m-%d')
df3 = df3[df3.date >= datetime.datetime(2010,1,1)].fillna(0)
df3['total'] = df3.apply(lambda row: row.roa + row.roe + row.gross + row.net, axis=1)
df3
```
