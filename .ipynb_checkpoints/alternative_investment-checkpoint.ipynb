```python
df_repay = df1.reset_index(drop=True).pivot(index=['date'], columns=['code'], values=['total']).dropna(axis=1)['total'].reset_index().set_index('date')
df_repay
```

```python
def change_extreme(row):
    m=row.mean()
    sd=row.std()
    row=row.apply(lambda x:x if m-3*sd<=x<=m+3*sd else (m+3*sd if x>m+3*sd else m-3*sd))
    return row
```

```python
df_repay = df_repay.apply(change_extreme, axis=1)
df_repay = df_repay.apply(lambda row: (row-row.mean())/row.std(), axis=1)
df_repay
```

```python
df_repay.min(axis=1)
```

```python
df_rank = df_repay.rank(axis=1, method='min')
df_rank
```

```python
def calcResults(w1, w2, w3):
    ff = factor1 * w1 + factor2 * w2 + factor3 * w3
    rank = ff.rank(axis=1, method='min')
    rank = rank.iloc[:-1, :]
    return df_return[rank>4400].mean(axis=1).mean()
```
